#### **4.** Supprimez la première ligne de données : nous ne pouvons pas savoir si le patient est sain ou malade (diagnostic = NAN). Cette ligne n'est donc pas utilisable.

**<span style="color:blue">mimm_data[-c(1), ]</span>** :<br>
Les crochets **<span style="color:blue">[ ]</span>** permettent de sélectionner des lignes ou des colonnes du tableau **<span style="color:blue">mimm_data</span>**.<br>
L'opération **<span style="color:blue">-c(1)</span>** permet de sélectionner le 1er élément d'une liste et de le soustraire via l'opérateur **<span style="color:blue">-</span>**.<br>
La virgule **<span style="color:blue">,</span>** est placée **après** l'opération de sélection et précise qu'on applique la sélection sur la liste des lignes (et non des colonnes auquel cas la virgule aurait été placée **avant**).<br>
Pour résumé **on soustrait la première ligne du tableau mimm_data**.<br>
<br>
**<span style="color:blue">mimm_data =</span>** :<br>
Cela permet d'attribuer à la variable **<span style="color:blue">mimm_data</span>** les nouveaux changements qui ont été opérés.

```r
mimm_data = mimm_data[-c(1), ]
```

#### **6.** Supprimez la colonne Gene_supprimer.

**<span style="color:blue">subset(mimm_data, select = -c(Gene_supprimer))</span>** <br>
La fonction **<span style="color:blue">subset()</span>** permet de récupérer une partie d'un tableau. On lui donne la variable **<span style="color:blue">mimm_data</span>** pour déterminer précisément dans quel tableau connu la partition va s'opérer.<br>
Le paramètre **<span style="color:blue">select =</span>** permet de déclarer ce que l'on veut sélectionner. Ici, via **<span style="color:blue">-c(Gene_supprimer)</span>** on précise qu'on veut enlever (par le **<span style="color:blue">-</span>** ) la colonne **<span style="color:blue">Gene_supprimer</span>**. Cette fois-ci on supprime un élément en utilisant son nom et non sa place (cf 4.) <br>
Pour résumé **on prend le tableau entier et on sélectionne la colonne Gene_supprimer comme étant à soustraire du tableau**.<br>
<br>
**<span style="color:blue">mimm_data =</span>** <br>
Cela permet à nouveau d'attribuer à la variable **<span style="color:blue">mimm_data</span>** les nouveaux changements qui ont été opérés.

```r
mimm_data = subset(mimm_data, select = -c(Gene_supprimer))
```

#### **7.** Renommez la colonne Gene_renommer en Gene_NEW.

**<span style="color:blue">rename(mimm_data,Gene_NEW = Gene_renommer)</span>** <br>
La fonction **<span style="color:blue">rename()</span>** permet de renommer une colonne de tableau. On lui donne la variable **<span style="color:blue">mimm_data</span>** pour déterminer précisément dans quel tableau connu ce changement va être réalisé.<br>
On peut changer le nom d'autant de colonne que l'on veut. Pour cela on donne le nouveau nom souhaité sans espace et caractère spécial. Ici on donne **<span style="color:blue">Gene_NEW</span>** et on doit définir quel est le nom de colonne existant correspondant dans le tableau, ici c'est **<span style="color:blue">Gene_renommer</span>**.

**<span style="color:blue">mimm_data =</span>** <br>
Cela permet à nouveau d'attribuer à la variable **<span style="color:blue">mimm_data</span>** les nouveaux changements qui ont été opérés.

```r
mimm_data = rename(mimm_data, Gene_NEW = Gene_renommer)
```

#### **9.** Observez l'expression du gène CDH3 (axe des ordonnées) chez les patients sains et les patients atteints d'un cancer (axe des abscisses).

**<span style="color:blue">ggplot(aes(x = Diagnostic, y = Gene_CDH3, ), data = mimm_data)</span>** <br>
La fonction **<span style="color:blue">ggplot()</span>** permet d'initialiser " un objet ggplot". Les objets ggplot sont utilisés dans R pour créer des figures parfois complexes.<br>
La fonction **<span style="color:blue">aes()</span>** sert à déterminer " l'esthétique " de la figure créée. En commençant par définir les colonnes du tableau qu'on veut voir sur l'axe des abscisses (avec **<span style="color:blue">x =</span>** ) et l'axe des ordonnées (avec **<span style="color:blue">y =</span>**). On précise dans quel tableau on va chercher ces colonnes avec **<span style="color:blue">data =</span>**. En allant plus loin on pourrait affiner le visuel de la figure en assignant par exemple des couleurs aux données représentées sur la figure.

**<span style="color:blue">+ geom_boxplot()</span>**<br>
Le **<span style="color:blue">+</span>** permet d'ajouter une nouvelle spécificité à l'objet ggplot.<br>
La fonction **<span style="color:blue">geom_boxplot()</span>** précise qu'on souhaite que la figure qui représente l'expression du gène CDH3 en fonction du diagnostic doit se faire sous forme d'un **boxplot**.


```r
ggplot(aes(y = Gene_CDH3,x = Diagnostic), data = mimm_data) + geom_boxplot()
```
